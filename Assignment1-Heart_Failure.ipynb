```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',30)
```

```python
df = pd.read_csv('heart-failure-dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['age'] = df['age'].astype(int)
```

```python
df.dtypes
```

```python
# What is the distribution of age among heart failure patients in the dataset
```

```python
sns.histplot(df['age'],kde=True) 
plt.title('Distribution of Age')
plt.show()                                 ## distribution of age among heart failure patients in the dataset
```

```python
# How does the death rate vary with age
```

```python
df.columns
```

```python
df['DEATH_EVENT'].unique()  # 1 for success i.e patient is dead 0 for failure i.e patient survived
```

```python
sns.countplot(x='DEATH_EVENT',hue='age',data = df,palette='rocket')
plt.xlabel('Death Event')
plt.ylabel('Count')
plt.title('Death rate vary with Age')      
plt.show()
```

```python
sns.scatterplot(x='age', y='DEATH_EVENT', data=df, hue='DEATH_EVENT', palette='rocket')
plt.title('Death Rate Variation with Age')
plt.xlabel('Age')
plt.ylabel('Death Event (1 = Death, 0 = No Death)')
plt.legend(title='Death Event', loc='upper right', labels=['No Death', 'Death'])
plt.grid(True)
plt.show()
```

```python
## ²´ What is the percentage of male and female patients in the dataset?
```

```python
df['sex'].value_counts().plot.pie(autopct='%1.1f%%')
plt.legend(title='sex',labels=['Male','Female'])
plt.title('sex distribution')
plt.show()
```

```python
## How does the platelets count vary among different age groups
```

```python
sns.histplot(x='age',y='platelets',data=df,kde=True)
plt.show()
```

```python
sns.scatterplot(x='age',y='platelets',data=df,palette='rocket')
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x='age', y='platelets', data=df, palette='viridis')
plt.title('Platelet Count Variation Across Age Groups')
plt.xlabel('Age')
plt.ylabel('Platelet Count')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# is there a correlation b/w cretanine and sodium levels in the blood?
```

```python
df.corr()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x='serum_creatinine', y='serum_sodium', data=df, color='blue', alpha=0.7)
plt.title('Creatinine vs. Sodium Levels in Blood')
plt.xlabel('Serum Creatinine')
plt.ylabel('Serum Sodium')
plt.grid(True)
plt.show()
```

```python
correlation_coefficient = df['serum_creatinine'].corr(df['serum_sodium'])
print(f"Pearson correlation coefficient: {correlation_coefficient}")
```

```python
# How does the prevalence of high blood pressure differ b/w male and female patients?
```
